```r
library("plyr")
library(descr)
library("partykit")
library("tidyverse")
library("ggplot2")
data <- read.csv(file="../input/hvm.csv")
```

```r
head(data)
```

```r
summary(data)
```

```r
str(data)
```

```r
ggplot(data,aes(mental,physical))+
geom_tile(aes(fill=satisfaction))
```

```r
ggplot(data,aes(status  ,satisfaction))+
geom_violin()
```

```r
linearMod <- lm(age ~ satisfaction, data=data)
print(linearMod)
```

```r
ggplot(data,aes(food  ,satisfaction))+
geom_jitter(aes(color=status,size=mental,shape=gender))+
geom_smooth(method = lm, se = FALSE, color='#ff9a7d')
```

```r
cor(data$age,data$satisfaction)
```
